```python
import pandas as pd
import yfinance as yf
```

```python
 from fbprophet import Prophet
```

```python
#"BBDC4.SA ITUB4.SA SANB4.SA BIDI4.SA BBAS3.SA ^BVSP"
acoes = ['ITUB4.SA']
carteira = pd.DataFrame()
for i in acoes:
  carteira[i] = yf.download(i, start="2021-01-01", end="2021-10-20")['Adj Close']
```

```python
df = pd.DataFrame()
```

```python
carteira
```

```python
df['y'] = carteira['ITUB4.SA']
df['ds'] = carteira.index
```

```python
df.head()
```

```python
modelo = Prophet(seasonality_mode='multiplicative')
```

```python
modelo.fit(df)
```

```python
futuro = modelo.make_future_dataframe(periods=10)
```

```python
futuro.tail()
```

```python
prever = modelo.predict(futuro)
```

```python
prever.tail()
```

```python
prever[["ds","yhat","yhat_lower","yhat_upper"]].head()
```
